### Line Plot

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(20,10)) 
ax[0].plot(df['Open']) 
ax[0].set_xlabel('Date') 
ax[0].set_ylabel('Opening Price') 
ax[0].set_title('Trend of Opening Price of Stocks') 

ax[1].plot(df['Close']) 
ax[1].set_xlabel('Date') 
ax[1].set_ylabel('Closing Price') 
ax[1].set_title('Trend of Closing Price of Stocks') 

plt.show()
```

```python
df['InsDate'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['InsDate'].dt.strftime('%Y')
df['Month'] = df['InsDate'].dt.strftime('%m')
df['Day'] = df['InsDate'].dt.strftime('%d')
```

```python
df.drop('InsDate',axis=1)
```

```python
df.head()
```

```python
#Yearly trends 
fig,ax=plt.subplots(nrows=1,ncols=4,figsize=(20,10)) 
ax[0].plot(df.groupby('Year')['Open'].sum()) 
ax[0].set_xlabel('Year') 
ax[0].set_ylabel('Opening Price') 
ax[0].set_title('Trend of Opening Price of Stocks') 

ax[1].plot(df.groupby('Year')['Close'].sum()) 
ax[1].set_xlabel('Year') 
ax[1].set_ylabel('Closing Price') 
ax[1].set_title('Trend of Closing Price of Stocks') 

ax[2].plot(df.groupby('Year')['High'].max()) 
ax[2].set_xlabel('Year') 
ax[2].set_ylabel('Opening Price') 
ax[2].set_title('Trend of Highest Price of Stocks') 

ax[3].plot(df.groupby('Year')['Open'].min()) 
ax[3].set_xlabel('Year') 
ax[3].set_ylabel('Opening Price') 
ax[3].set_title('Trend of Lowest Price of Stocks') 

plt.show()
```

```python
#CORRELATION
heatmap = sns.heatmap(df.corr(), cmap='BuPu')
plt.figure(figsize=(20,10))
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```

```python
#Features Correlation - Open
heatmap = sns.heatmap(df.corr()[['Open']].sort_values(by='Open', ascending=False), vmin=-1, vmax=1, annot=True, cmap='Blues')
heatmap.set_title('Features Correlating with Opening Price', fontdict={'fontsize':18}, pad=16);
```

```python
#Features Correlation - High
heatmap = sns.heatmap(df.corr()[['High']].sort_values(by='High', ascending=False), vmin=-1, vmax=1, annot=True, cmap='Greens')
heatmap.set_title('Features Correlating with Highest Price of the day', fontdict={'fontsize':18}, pad=16);
```

```python
#Features Correlation - Low
heatmap = sns.heatmap(df.corr()[['Low']].sort_values(by='Low', ascending=False), vmin=-1, vmax=1, annot=True, cmap='YlGnBu')
heatmap.set_title('Features Correlating with Lowest Price of the Day', fontdict={'fontsize':18}, pad=16);
```

```python
#Features Correlation - Close
heatmap = sns.heatmap(df.corr()[['Close']].sort_values(by='Close', ascending=False), vmin=-1, vmax=1, annot=True, cmap='PiYG')
heatmap.set_title('Features Correlating with Closing Price', fontdict={'fontsize':18}, pad=16);
```

```python
#Features Correlation - Volume
heatmap = sns.heatmap(df.corr()[['Volume']].sort_values(by='Volume', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BuPu')
heatmap.set_title('Features Correlating with Volume of stocks sold in a day', fontdict={'fontsize':18}, pad=16);
```
